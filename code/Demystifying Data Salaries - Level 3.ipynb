## 💾 The data

The data comes from a survey hosted by an HR consultancy, available in `'salaries.csv'`.

#### Each row represents a single employee's salary record for a given year:
- **`work_year`** - The year the salary was paid.  
- **`experience_level`** - Employee experience level:  
  - **`EN`**: Entry-level / Junior  
  - **`MI`**: Mid-level / Intermediate  
  - **`SE`**: Senior / Expert  
  - **`EX`**: Executive / Director  
- **`employment_type`** - Employment type:  
  - **`PT`**: Part-time  
  - **`FT`**: Full-time  
  - **`CT`**: Contract  
  - **`FL`**: Freelance  
- **`job_title`** - The job title during the year.  
- **`salary`** - Gross salary paid (in local currency).  
- **`salary_currency`** - Salary currency (ISO 4217 code).  
- **`salary_in_usd`** - Salary converted to USD using average yearly FX rate.  
- **`employee_residence`** - Employee's primary country of residence (ISO 3166 code).  
- **`remote_ratio`** - Percentage of remote work:  
  - **`0`**: No remote work (<20%)  
  - **`50`**: Hybrid (50%)  
  - **`100`**: Fully remote (>80%)  
- **`company_location`** - Employer's main office location (ISO 3166 code).  
- **`company_size`** - Company size:  
  - **`S`**: Small (<50 employees)  
  - **`M`**: Medium (50–250 employees)  
  - **`L`**: Large (>250 employees)

### Section Q2
Predictive models were constructed using employees' experience level, company location, & remote ratio to estimate an employee's salary. The importances of each feature were evaluated.

### Employee Residence
Examine employees' average & maximum salaries per "employee_residence".

```python
#Filter the data for the specified jobs
q1_data_I = salaries_df[salaries_df.job_title.isin(['Data Analyst','Data Scientist',
                                                        'Machine Learning Engineer'])].reset_index(drop=True)

#Obtain the name of each country in "employee_residence"
q1_data_I['employee_res_name'] = [pycountry.countries.get(alpha_2=x).name for x in q1_data_I.employee_residence]
#obtain overall avg salary across these 3 jobs
display("Average salary of these three jobs: ${}.".format(round(q1_data_I.salary_in_usd.mean(), 2)))

#Inspect values of the "employee_residence", variable
q1_data_res_I = q1_data_I.employee_residence.value_counts().reset_index()
#display(q1_data_res_I)

#Analyze salaries per "employee_residence"
q1_data_res_II = q1_data_I.groupby('employee_res_name').salary_in_usd.describe()

#adjust column values
q1_data_res_II.iloc[:,2:] = round(q1_data_res_II.iloc[:,2:], 2)
q1_data_res_II['count'] = q1_data_res_II['count'].astype('int')
q1_data_res_II = q1_data_res_II.sort_values('mean', ascending=False).reset_index()
display(q1_data_res_II)

#Plot avg salary per "employee_residence"
plt.figure(figsize=(9,5))
ax1 = sns.barplot(data=q1_data_res_II[:10], x='employee_res_name', y='mean', color=sns.color_palette()[0])
#annotate each bar w/# records
ax1.bar_label(ax1.containers[-1], labels=[f"{num}" for num in q1_data_res_II['count'][:10]], fontsize=8.5)
#edit plot details
plt.title('Average Employee Salary & Their Country of Residence (2020-2024)')
plt.xlabel('Country')
plt.xticks(fontsize=9, rotation=10)
plt.ylabel('Average Salary ($)')
plt.yticks(ticks=range(0,161000,20000))
plt.grid(axis='y',alpha=0.3,color='gray')
plt.show();
```

About 90% of data analysts, data scientists, & machine learning engineers in this dataset reside in the United States (US). Another six percent are from Canada (CA) the United Kingdom (GB). Of the 70 unique employee residence values, there are 14 countries that only have one record each.

The plot above displays the ten largest average salaries of employees based on their country of residence. Specifically, these employees are either data analysts, data scientists, or machine learning engineers. These averages span the whole dataset from 2020-2024. The numbers directly above each bar indicate the number of employee records that pertained to each respective country.

Of these three jobs, employees from Mexico earn the highest average salary; about \\$170,000. Recall that most employees that work one of these three jobs are from the United States. These employees make about \\$160,000 on average over these five years. For reference, data analysts, data scientists, & machine learning engineers make about \\$154,000 on average across 2020-2024. Only employees from three countries--Mexico, Puerto Rico, & the US--make a greater average salary than this \$154,000.

As opposed to linear regression, no scaling is required for tree-based models. Nonetheless, categorical features still need to be encoded to numerical values.

A random forest model is composed of multiple decision tree models. More specifically, a tree is an individual decision tree model that is trained on a slightly different subset of data than all other trees. Final predictions are made by compiling predictions across all trees.

One primary hyperparameter for random forest regressor models is the "n_estimators" parameter, which specifies the number of trees to use in a random forest model. This hyperparameter was tuned in isolation to evaluate how it affected the model's performance. Later, this parameter was also used for more tuning but in combination with other parameters. In regards to feature importance, it can essentially be described as how much this feature is used in each tree of the forest. So, larger feature importance coefficients indicate that a feature variable is more important in predicting the target variable in a random forest model.

```python
#assess different n_estimators values of a RF model on the training data. make sure to set the random_state
scores = []
for n in [10, 20, 50, 100, 200]:
    #instantiate the model with different # trees
    rf = RandomForestRegressor(n_estimators=n, random_state=40)
    #fit the model to the training data
    rf.fit(X_train, y_train)
    #compile the R^2 scores for each model
    scores.append(rf.score(X_test, y_test))

display(scores)
```
