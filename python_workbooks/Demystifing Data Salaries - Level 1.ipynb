## 📖 Background
You work for an international HR consultancy helping companies attract and retain top talent in the competitive tech industry. As part of your services, you provide clients with insights into industry salary trends to ensure they remain competitive in hiring and compensation practices.

Your team wants to use a data-driven approach to analyse how various factors—such as job role, experience level, remote work, and company size—impact salaries globally. By understanding these trends, you can advise clients on offering competitive packages to attract the best talent.

In this competition, you’ll explore and visualise salary data from thousands of employees worldwide. If you're tackling the advanced level, you'll go a step further—building predictive models to uncover key salary drivers and providing insights on how to enhance future data collection.

## 💾 The data

The data comes from a survey hosted by an HR consultancy, available in `'salaries.csv'`.

#### Each row represents a single employee's salary record for a given year:
- **`work_year`** - The year the salary was paid.  
- **`experience_level`** - Employee experience level:  
  - **`EN`**: Entry-level / Junior  
  - **`MI`**: Mid-level / Intermediate  
  - **`SE`**: Senior / Expert  
  - **`EX`**: Executive / Director  
- **`employment_type`** - Employment type:  
  - **`PT`**: Part-time  
  - **`FT`**: Full-time  
  - **`CT`**: Contract  
  - **`FL`**: Freelance  
- **`job_title`** - The job title during the year.  
- **`salary`** - Gross salary paid (in local currency).  
- **`salary_currency`** - Salary currency (ISO 4217 code).  
- **`salary_in_usd`** - Salary converted to USD using average yearly FX rate.  
- **`employee_residence`** - Employee's primary country of residence (ISO 3166 code).  
- **`remote_ratio`** - Percentage of remote work:  
  - **`0`**: No remote work (<20%)  
  - **`50`**: Hybrid (50%)  
  - **`100`**: Fully remote (>80%)  
- **`company_location`** - Employer's main office location (ISO 3166 code).  
- **`company_size`** - Company size:  
  - **`S`**: Small (<50 employees)  
  - **`M`**: Medium (50–250 employees)  
  - **`L`**: Large (>250 employees)

# Q1
How many records are in the dataset, and what is the range of years covered?

**Plan of action:**
- The number of records in the dataset was stated in the cell previous using `.info()`; 57,194 rows.
- Use the "work_year" variable to explore the range of years in the dataset.

```python
#Explore the "work_year" variable
#count # records per year
records_per_year = salaries_df.work_year.value_counts(dropna=False).reset_index().sort_values('work_year').reset_index(drop=True)
#add a column indicating the proportion of records per year
records_per_year['percentage'] = round((records_per_year['count']/salaries_df.shape[0])*100, 1)
display(records_per_year)

#visualize the data
ax1 = sns.countplot(data=salaries_df, x='work_year')
#label each bar with the % of total records
ax1.bar_label(ax1.containers[-1], labels=[f"{num} %" for num in records_per_year['percentage']])
#edit plot details
plt.title('# Records per Year')
plt.xlabel('Year')
plt.ylabel('# Records')
plt.show();
```

The plot above displays the number of records for each unique year a salary was paid. The data spans 2020-2024. The percentages above each bar indicate how many observations of the entire dataset each year accounts for.

From this plot, the number of records appears to increase exponentially on a yearly basis. Recall that the data comes from a survey hosted by an HR consultancy, & each record represents a single employee's salary record for a given year. As such, this suggests that the survey that supplies the dataset compiled exponentially more employee salary records over these five years. Alternatively, the HR consultancy that hosted this survey may have obtained access to exponentially more employees over time who were willing to answer their survey.

# Q2
What is the average salary (in USD) for Data Scientists and Data Engineers? Which role earns more on average?

**Plan of action:**
- Use the "job_title" variable to filter for data scientists & data engineers. Use the "salary_in_usd" column to examine their salaries.

```python
#Explore the different jobs in the dataset
num_jobs = salaries_df.job_title.value_counts().reset_index()
num_jobs['percentage'] = num_jobs['count']/salaries_df.shape[0]
display(num_jobs.head(5))

#Explore employee salaries (USD)
display(salaries_df.salary_in_usd.describe().reset_index().T)
display("Median salary: ${}.".format(salaries_df.salary_in_usd.median()))
```

Some observations of the "job_title" & "salary_in_usd" variables.
- Data scientists & data engineers are the two most popular jobs in the dataset. Together, they account for about 30% of the data points.
- The values of employees salaries in USD range from \$15,000-\$800,000. The average & median salary is about \$159,000 & \$149,000 respectively.

```python
#filter for the jobs of interest
data_sci_eng = salaries_df[salaries_df.job_title.isin(['Data Scientist', 'Data Engineer'])]
#need to categorize the "job_title" variable to make sure plotting is consistent
data_sci_eng['job_title'] = pd.Categorical(data_sci_eng.job_title, categories=['Data Scientist','Data Engineer'], ordered=True)
#calculate salary metrics for these jobs
q2_data = data_sci_eng.groupby('job_title')['salary_in_usd'].describe()
#reorganize the data
q2_data = q2_data.T.reset_index().melt(id_vars='index', value_vars=['Data Engineer','Data Scientist']).rename(
                columns={'index':'metric', 'value':'salary_in_usd', 'level_0':'index'})
#round the salary values
q2_data['salary_in_usd'] = round(q2_data['salary_in_usd'], 2)
display(q2_data.sort_values('metric').reset_index(drop=True).iloc[6:,:])
```

```python
#Visualize the data
ax1 = sns.boxplot(data=data_sci_eng, x='job_title', y=data_sci_eng['salary_in_usd']/100000, fill=False)
#add the mean salary
sns.scatterplot(data=q2_data[q2_data.metric == 'mean'], x='job_title', y=q2_data['salary_in_usd']/100000, marker='*', color='red', s=120)

#edit plot details
plt.title('Salaries of Data Scientists & Data Engineers')
plt.xlabel('Job')
plt.ylabel('Salary ($100,000)')
plt.show();
```

The figure above illustrates the distribution of salaries of data scientists & data engineers (in USD) using box plots. Note that the vertical axis is in units of \$100,000. In a box plot, there are several metrics displayed. From bottom to top of the two box plots above:
- The bottom-most line displays the lower extreme of the vertical axis,
- The bottom line of the closed box displays the lower quartile (25th percentile),
- The middle line of the closed box displays the middle quartile or median (50th percentile),
- The upper line of the closed box displays the upper quartile (75th percentile),
- The upper-most line displays the upper extreme,
- The points represent outliers.
- The closed box highlights most of the data.

The red stars were superposed on top of these box plots to indicate the mean salary for each job. Overall, it appears that salaries of data scientists are slightly greater than those of data engineers. In fact, the average salary of data scientists is about $10,000 more than for data engineers.

Recall that across all jobs in the dataset, the average salary is about \$159,000. As such, data scientists make about the average salary of individuals in this dataset, whereas data engineers make slightly less than average.

```python
#Analyze the salaries of this subset of data over time
#count # such records per year
data_sci_eng_nums = data_sci_eng.groupby(['work_year','job_title']).size().reset_index().rename(columns={0:'num'})
#display(data_sci_eng_nums)
#get avg salary per year
q2_yr_avgs = data_sci_eng.groupby(['work_year','job_title'])['salary_in_usd'].mean().reset_index()
q2_yr_avgs['salary_in_usd'] = round(q2_yr_avgs['salary_in_usd'], 2)
display(q2_yr_avgs)

#visualize the data
fig, axes = plt.subplots(nrows=2, figsize=(9,9))
ax1 = sns.boxplot(data=data_sci_eng, x='work_year', y=data_sci_eng['salary_in_usd']/100000, hue='job_title', fill=False, gap=0.1, ax=axes[0])
#plot the yearly means
ax2 = sns.lineplot(data=q2_yr_avgs, x='work_year', y=q2_yr_avgs['salary_in_usd']/100000, hue='job_title', marker='o', alpha=0.7, ax=axes[1])

#edit plot details
fig.suptitle('Yearly Salaries of Data Scientists & Data Engineers')
for i in [0,1]:
    plt.sca(axes[i])
    plt.xlabel('Year')
    plt.ylabel('Salary ($100,000)')
    plt.legend(title='Job', fontsize=8)
    #for the 2nd plot specifically:
    if i==1:
        plt.xticks(ticks=range(2020,2025,1))
        plt.ylabel("Avg. Salary ($100,000)")
    plt.show();
```

The two plots above illustrate the distributions of employee salaries over time. Specifically, this data pertains to data engineers & data scientists between 2020-2024. The top plot shows the distribution of these employees' salaries, whereas the bottom plot displays the average salary per year. Note that the vertical axes of these plots, which are in units of \$100,000, are not the same. Specifically, that of the latter plot is much smaller. Also, note that the number of data points per year differ significantly; there are only about 40 in 2020 & 13,000 in 2024. On the other hand, between the two jobs, the yearly number of records is somewhat equivalent.

From the first plot, the distributions of salaries across the two jobs are relatively similar, but they appear to increase in disparity in the last two years. Additionally, the distribution of salaries of data scientists appears to be larger in value than for data engineers in four of the five years (2020, 2022-2024).

In the second plot, the average yearly salary is greater for data scientists in three of the five years (2020, 2023-2024). In the two years where data engineers earned more on average, it is by a very slight margin. The average salary for data engineers increased every year. For data scientists, their salary was at its lowest in 2021 & it peaked in 2023.

# Q3
How many full-time employees based in the US work 100% remotely?

**Plan of action:**
- Use the "employment_type" variable to filter for full-time employees. Use the "employee_residence" variable to search for employees based in the US. Use the "remote_ratio" variable to find employees who work remotely.

```python
#Explore the three variables of relevance
#"employment_type"
num_emp_type = salaries_df.groupby('employment_type').size().reset_index().rename(columns={0:'num'})
num_emp_type['percentage'] = round(100*num_emp_type['num']/salaries_df.shape[0],2)
display(num_emp_type)
#"employee_residence"
num_emp_res = salaries_df.groupby('employee_residence').size().reset_index().rename(columns={0:'num'})
num_emp_res['percentage'] = round(100*num_emp_res['num']/salaries_df.shape[0],2)
display(num_emp_res[num_emp_res.employee_residence == "US"])
#"remote_ratio"
num_remote = salaries_df.groupby('remote_ratio').size().reset_index().rename(columns={0:'num'})
num_remote['percentage'] = round(100*num_remote['num']/salaries_df.shape[0],2)
display(num_remote)
```

Some observations of the three variables of interest.
- **employment_type:**
    - About 99.6% of employees in the dataset work full-time.
- **employee_residence:**
    - About 90.7% of employees in the dataset are primarily based in the US.
- **remote_ratio:**
    - About 22.0% of employees in the dataset work entirely remotely.

```python
#Answer Q3
q3_data = salaries_df[(salaries_df.employment_type == "FT") & (salaries_df.employee_residence == "US") & (salaries_df.remote_ratio == 100)]
display("Of the 57,194 records in the dataset, {} of them pertain to employees who work full-time, are based in the US, & work completely remotely; about {}%.".format(q3_data.shape[0], round(100*q3_data.shape[0]/salaries_df.shape[0], 1)))

#analyze the salaries of these individuals
display(round(q3_data['salary_in_usd'].describe(), 2).reset_index().T)
```

```python
#Analyze the salaries of this subset of data over time
#count # such records per year
q3_data_nums = q3_data.groupby('work_year').size().reset_index().rename(columns={0:'num'})
#display(q3_data_nums)
#get avg salary per year
q3_yr_avgs = q3_data.groupby('work_year')['salary_in_usd'].mean().reset_index()
q3_yr_avgs['salary_in_usd'] = round(q3_yr_avgs['salary_in_usd'], 2)
display(q3_yr_avgs)

#plot salary distributions over time
fig, axes = plt.subplots(nrows=2, figsize=(8,9))
ax1 = sns.boxplot(data=q3_data, x='work_year', y=q3_data['salary_in_usd']/100000, fill=False, ax=axes[0])
#add the yearly means
ax2 = sns.lineplot(data=q3_yr_avgs, x='work_year', y=q3_yr_avgs['salary_in_usd']/100000, marker='o', color='red', ax=axes[1])

#edit plot details
fig.suptitle('Yearly Salaries of Full-time, US-based, Fully Remote Employees')
for i in [0,1]:
    plt.sca(axes[i])
    plt.xlabel('Year')
    plt.ylabel('Salary ($100,000)')
    #for the 2nd plot specifically:
    if i==1:
        plt.xticks(ticks=range(2020,2025,1))
        plt.ylabel('Avg Salary ($100,000)');
```

The plots above illustrate the distributions of employee salaries over time. Specifically, this data pertains to full-time, US-based, fully remote employees between 2020-2024. The top plot shows the distribution of these employees' salaries, whereas the bottom plot displays the average salary per year. Note that the vertical axes of these plots, which are in units of $100,000, are not the same. Specifically, that of the latter plot is much smaller. Also, note that the number of data points per year differ significantly; there are only 20 in 2020 & over 8,000 in 2024.

The average salary of these records is greatest in 2023. In the most recent year, 2024, this average fell by a couple of hundred of US dollars. Overall, this number has remained fairly stagnant apart from a ~\$10,000 dip in 2021. Across these five years, the average salary is about \$155,000, which is only exceeded in the two most recent years. Each of these yearly salaries is below the average salary across the entire dataset--about \$159,000.
