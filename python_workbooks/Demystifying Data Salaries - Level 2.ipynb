## 💾 The data

The data comes from a survey hosted by an HR consultancy, available in `'salaries.csv'`.

#### Each row represents a single employee's salary record for a given year:
- **`work_year`** - The year the salary was paid.  
- **`experience_level`** - Employee experience level:  
  - **`EN`**: Entry-level / Junior  
  - **`MI`**: Mid-level / Intermediate  
  - **`SE`**: Senior / Expert  
  - **`EX`**: Executive / Director  
- **`employment_type`** - Employment type:  
  - **`PT`**: Part-time  
  - **`FT`**: Full-time  
  - **`CT`**: Contract  
  - **`FL`**: Freelance  
- **`job_title`** - The job title during the year.  
- **`salary`** - Gross salary paid (in local currency).  
- **`salary_currency`** - Salary currency (ISO 4217 code).  
- **`salary_in_usd`** - Salary converted to USD using average yearly FX rate.  
- **`employee_residence`** - Employee's primary country of residence (ISO 3166 code).  
- **`remote_ratio`** - Percentage of remote work:  
  - **`0`**: No remote work (<20%)  
  - **`50`**: Hybrid (50%)  
  - **`100`**: Fully remote (>80%)  
- **`company_location`** - Employer's main office location (ISO 3166 code).  
- **`company_size`** - Company size:  
  - **`S`**: Small (<50 employees)  
  - **`M`**: Medium (50–250 employees)  
  - **`L`**: Large (>250 employees)

```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#import the data & inspect it
salaries_df = pd.read_csv('salaries.csv')
display(salaries_df.head(5))
display(salaries_df.info())
```

In the original dataset, there are 57,194 observations & 11 columns. There are no missing values.

# Q1
Create a bar chart displaying the top 5 job titles with the highest average salary (in USD).

**Plan of action:**
- Use the "job_title" & "salary_in_usd" variables.
- It may be valuable to evaluate the number of records each job accounts for.

```python
#obtain summary statistics of "salary_in_usd" per "job_title"
q1_data_I = salaries_df.groupby('job_title').salary_in_usd.describe().reset_index()
#round salaries to 2 decimals
q1_data_I.iloc[:,2:] = round(q1_data_I.iloc[:,2:], 2)
#convert the "count"s to integers
q1_data_I['count'] = q1_data_I['count'].astype('int')

display(q1_data_I.sort_values('count', ascending=False).reset_index(drop=True))
#inspect distribution of job frequencies
display(q1_data_I['count'].describe())
```

In the dataset, there are 253 unique job titles. 20 of them only account for one data point, whereas the most popular job, data scientists, account for 9,706 records. Most of the jobs that are present account for between three & 50 observations in the dataset each. This information is prevalent because some jobs with few observations may act as outliers when determining which have the largest average salaries. On the other hand, metrics of jobs with many observations can be diluted & less prone to influence from outliers. Without normalized data, calculating aggregate statistics like an average can involve biases.

To avoid this, a random sampling process could be utilized, but when some jobs only have one data point, this would still have its issues. The best form of this process is beyond the scope of this project, but a small example could still be evaluated which could consist of sampling a smaller subset of the main dataset & finding the job titles with the highest salaries. Given the size of the original dataset (\~57,000 rows), such a subset could be 20% (\~11,400) of the original dataset.

As another alternative, jobs with too few observations could be ignored from this process so that their presence, if they are outliers, cannot influence analysis; however, it should be noted that outliers do exist in many real-world datasets & applications. In this case, given the number of records per job in this dataset, a better range would include jobs that have at least three records--the 25th percentile of counts of "job_title"s. Of the 253 unique jobs, 56 of them have less than three unique data points (\~22%). However, only 20 have one data point a piece, so jobs that have at least two records might be a better restriction than three.

For simplicity, two processes were evaluated:
1) Find the jobs with the top average salaries across the entire dataset.
2) Of jobs with at least two records, find the jobs with the top average salaries across the entire dataset.
3) Find the jobs with the top average salaries across a subset of the dataset, the sample size of which is 20%.

```python
#1) Find 5 jobs with highest avg salaries -- entire dataset
q1_data_II = q1_data_I.sort_values('mean', ascending=False).reset_index(drop=True)[:5]
display(q1_data_II)

#plot the data
plt.figure(figsize=(10,5))
ax1 = sns.barplot(data=q1_data_II, x='job_title', y=q1_data_II['mean']/100000, color=sns.color_palette()[0])
#label # records per job
ax1.bar_label(ax1.containers[-1], labels=[f"{num} record(s)" for num in q1_data_II['count']], fontsize=8.5)
#edit plot details
plt.title('Average Job Salary - All Jobs')
plt.xlabel('Job')
plt.xticks(fontsize=8, rotation=5)
plt.ylabel('Average Salary ($100,000)')
plt.yticks(fontsize=9)
plt.grid(axis='y',alpha=0.3,color='gray')
plt.show();
```

The plot above displays the average salary of five jobs in the tech industry. These five jobs have the highest average salaries across the whole dataset. The labels above each bar indicate the number of records each job accounts for in the dataset. Note that the vertical axis is in units of \\$100,000. For reference, the average salary across the whole dataset is about \$159,000 (from _Demystifying Data Salaries Level 1_).

Two of these highly paid jobs involve engineering managers. Two others have to do with machine learning (ML). Four of these jobs are positions that oversee the work of others; managers, leads, heads.

As discussed earlier, jobs with only a handful of observations may potentially pose as outliers. With more data points, outlier salaries are more likely to be diluted & regress to a more meaningful average. As such, the average salaries of the first two jobs--analytics engineering manager & data science tech lead--aren't as credible or representative because they only have one data point each. Specifically, these two jobs have an average salary that is nearly \$800,000 more than any other job, further suggesting the potentially unrepresentative nature of these data points. The last two jobs, head of machine learning & engineering manager, have the most credible data points of these five jobs.

```python
#2) Find 5 jobs w/highest avg salaries -- jobs with 2+ records
q1_data_III = q1_data_I[q1_data_I['count'] > 1].sort_values('mean', ascending=False).reset_index(drop=True)[:5]
display(q1_data_III)

#plot the data
plt.figure(figsize=(10,5))
ax1 = sns.barplot(data=q1_data_III, x='job_title', y=q1_data_III['mean']/100000, color=sns.color_palette()[0])
#label # records per job
ax1.bar_label(ax1.containers[-1], labels=[f"{num} record(s)" for num in q1_data_III['count']], fontsize=8.5)
#edit plot details
plt.title('Average Job Salary - Jobs with 2+ Records')
plt.xlabel('Job')
plt.xticks(fontsize=8, rotation=5)
plt.ylabel('Average Salary ($100,000)')
#use the same y-axis scale as in the previous plot for consistent comparison
plt.yticks(ticks=np.arange(0,4.1,0.5), fontsize=9)
ax1.set(ylim=(0,4.2))
plt.grid(axis='y',alpha=0.3,color='gray')
plt.show();
```

The plot above displays the average salary of five jobs in the tech industry. These five jobs have the highest average salaries of jobs that have at least two records in the original dataset. The labels above each bar indicate the number of records each job has. Note that the vertical axis is in units of \\$100,000.

The three highest paid jobs in this plot are also present in the previous plot. Two of these jobs are associated with artificial intelligence (AI), three of them associate with machine learning, & two of them associate with engineering. Moreover, four of these jobs are positions that oversee the work of others; leads, heads, managers.

Although jobs with only one data point have been removed from this plot, two of these five jobs only have two data points. While their statistics are more representative than jobs with only one data point, statistics of two records are still prone to being considerably influenced by outliers. As such, the analytics of the applied AI ML leads & machine learning performance engineers can be considered more lightly than those of the other three jobs.

If jobs with more data points are taken more seriously than those with only a couple, the average salary of machine learning heads stands out considerably in comparison to all others; by about \$25,000.

```python
#3) Find 5 jobs w/highest avg salaries -- sample 20% of the dataset

#sample 20% of the original the dataset, calculate avg salary per job
q1_data_IV = salaries_df.sample(frac=.2, random_state=10).reset_index(drop=True).groupby(
    'job_title').salary_in_usd.describe().reset_index()
#round salaries to 2 decimals
q1_data_IV.iloc[:,2:] = round(q1_data_IV.iloc[:,2:], 2)
#convert the "count"s to integers
q1_data_IV['count'] = q1_data_IV['count'].astype('int')

#obtain top 5 job-salary pairs
q1_data_V = q1_data_IV.sort_values('mean',ascending=False)[:5].reset_index(drop=True)
#the current "count" column displays the # records per job in the sampled data, but want the # records in the original dataset
    #can use the "q1_data_I" variable
    #make sure to differentiate between the two "count" columns
q1_data_V = q1_data_V.merge(q1_data_I[['job_title','count']], on='job_title', how='inner', suffixes=("_sample", "_all"))
display(q1_data_V)

#plot the data
plt.figure(figsize=(10,5))
ax1 = sns.barplot(data=q1_data_V, x='job_title', y=q1_data_V['mean']/100000, color=sns.color_palette()[0])
#label # records per job
ax1.bar_label(ax1.containers[-1], labels=[f"{num} record(s)" for num in q1_data_V['count_all']], fontsize=8.5)
#edit plot details
plt.title('Average Job Salary - Sample Subset')
plt.xlabel('Job')
plt.xticks(fontsize=8, rotation=10)
plt.ylabel('Average Salary ($100,000)')
#use the same y-axis scale as in the previous plot for consistent comparison
plt.yticks(ticks=np.arange(0,4.1,0.5), fontsize=9)
ax1.set(ylim=(0,4.2))
plt.grid(axis='y',alpha=0.3,color='gray')
plt.show();
```

The plot above displays the average salary of five jobs in the tech industry. The data that built this plot is from a sampled subset of the original dataset, specifically 20%. These five jobs have the highest average salaries of jobs in this subset. The labels above each bar indicate the number of records each job has in the **original** dataset, not the subsetted one. Note that the vertical axis is in units of \\$100,000.

Note that the highest paid jobs in this plot are both present in the previous two plots, & are amongst the highest paid jobs across the entire dataset on average. Four of these jobs relate to machine learning, & two of them associate with engineering. Three of these jobs are positions that oversee the work of others; heads, managers, directors.

Each of these five jobs has at least nine records in the original dataset. As such, their analytics can all be considered to be credible & representative. While the sampling process that produced this figure was elementary, it can still be informative for the purposes of this report.

The plot above displays the average salary of the three different kinds of remote-work types; no remote work, hybrid, or fully remote.

As such, it seems that companies value employees who work in person the most highly. It should be noted that there are considerably fewer hybrid records in the dataset than remote or non-remote workers. Perhaps hybrid work is not yet common enough for such employees to be valued at the same level. Analyses over time could help explore this concept.

The plot above displays the annual average salary of the three different kinds of remote-work types. It expands upon the most previous plot which explored the average salary for each remote-work type over the entire timeframe of the data.

As such, is is quite evident that in-person workers became increasingly more valuable over these five years. Fully remote workers also saw their work increase in value, but not to the same extent. Hybrid workers, on the other hand, saw a decrease in their value from 2020-2024.
